## CRUDE OIL PRODUCTION IN G7 COUNTRIES

#### 1.0 COUNTRY DATA CATEGORIZATION

```python
#CREATING A FUNCTION THAT CAN BE USED TO EXTRACT  DATA FROM A PARTICULAR G7 COUNTRY) 

def choose_country (country_name):
    country_data = DROP_G7[DROP_G7.country_or_area.isin(country_name)].sort_values("year").reset_index()
    country_data.drop ('index', axis=1 , inplace = True)
    return  country_data
```

```python
# GENERATING AND ASSIGNING DATA TO EACH COUNTRY IN G7

france_data = choose_country (["france"])
japan_data = choose_country (["japan"])
uk_data = choose_country (["united kingdom"])
us_data = choose_country (["united states"])
germany_data = choose_country (["germany"])
italy_data = choose_country (["italy"])
canada_data = choose_country (["canada"])
```

```python
france_data
```

#### 2.0 CRUDE OIL DATA CATEGORIZATION BASED ON COUNTRY

```python
#CREATING ANOTHER FUNCTION THAT CAN BE USED TO EXTRACT CATEGORY DATA FROM A PARTICULAR G7 COUNTRY) 

def choose_category (country_data, category_name):
    category_data = country_data[country_data.category.isin(category_name)].sort_values("year").reset_index()
    category_data.drop ('index', axis=1 , inplace = True)
    return  category_data

```

```python
france_crude = choose_category (france_data,["conventional_crude_oil"])
japan_crude = choose_category (japan_data,["conventional_crude_oil"])
uk_crude = choose_category (uk_data,["conventional_crude_oil"])
us_crude = choose_category (us_data,["conventional_crude_oil"])
germany_crude = choose_category (germany_data,["conventional_crude_oil"])
italy_crude = choose_category (italy_data,["conventional_crude_oil"])
canada_crude = choose_category (canada_data,["conventional_crude_oil"])
```

### 2.0.1 FURTHER FILTERING INTO CRUDEOIL PRODUCTION

```python
def choose_commodity (category_data, commodity_name):
    commodity_data = category_data[category_data.commodity_transaction.isin(commodity_name)].sort_values("year").reset_index()
    commodity_data.drop ('index', axis=1 , inplace = True)
    return  commodity_data
```

```python
france_crude_prod = choose_commodity (france_crude, ["conventional crude oil - production"])
japan_crude_prod = choose_commodity (japan_crude, ["conventional crude oil - production"])
uk_crude_prod = choose_commodity (uk_crude, ["conventional crude oil - production"])
us_crude_prod = choose_commodity (us_crude, ["conventional crude oil - production"])
germany_crude_prod = choose_commodity (germany_crude, ["conventional crude oil - production"])
italy_crude_prod = choose_commodity (italy_crude, ["conventional crude oil - production"])
canada_crude_prod = choose_commodity (canada_crude, ["conventional crude oil - production"])
```

```python
plt.figure(figsize = (15, 10))

x1 = france_crude_prod.year
y1 = france_crude_prod.quantity
plt.plot (x1, y1, label = "France")

x2 = japan_crude_prod.year
y2 = japan_crude_prod.quantity
plt.plot (x2, y2, label = "Japan")


x3 = uk_crude_prod.year
y3 = uk_crude_prod.quantity
plt.plot (x3, y3, label = "United Kingdom")

x4 = us_crude_prod.year
y4 = us_crude_prod.quantity
plt.plot (x4, y4, label = "United States")

x5 = germany_crude_prod.year
y5 = germany_crude_prod.quantity
plt.plot (x5, y5, label = "Germany")

x6 = italy_crude_prod.year
y6 = italy_crude_prod.quantity
plt.plot (x6, y6, label = "Italy")

x7 = canada_crude_prod.year
y7 = canada_crude_prod.quantity
plt.plot (x7, y7, label = "Canada")

plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel ("Year")
plt.ylabel ("Conventional Crude Oil production- (Metric tons, Thousand)")
plt.title ("Crude oil production for G7 countries", fontweight ="bold")
plt.legend(loc='best')
plt.show()
```
